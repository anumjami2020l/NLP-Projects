```python
class SubmitFormAction(Action):
    def name(self) -> Text:
        return "action_submit_form"

    def run(self, dispatcher: CollectingDispatcher, tracker: Tracker, domain: Dict[Text, Any]) -> List[Dict[Text, Any]]:
        # Access slot values after form submission
        name = tracker.get_slot("user_name")
        appointment_date = tracker.get_slot("appointment_date")
        service_type = tracker.get_slot("service_type")

        # Use the collected data for further actions or responses
        response = f"Thank you, {name}! Your appointment for {service_type} on {appointment_date} has been confirmed."
        dispatcher.utter_message(response)

        return []
```
